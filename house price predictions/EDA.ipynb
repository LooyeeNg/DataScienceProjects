### Column Types

```python
# list of numerical variables
numerical_features = [feature for feature in df_train.columns if df_train[feature].dtypes != 'O']
print('Total of numerical variables: ', len(numerical_features))
```

```python
# list of variables that contain year information
year_features = [feature for feature in df_train if 'Yr' in feature or 'Year' in feature]
print("Year features in Numerical Variables Count: {} / {}".format(len(year_features),len(numerical_features)))

# year_features
```

```python
discrete_features=[feature for feature in numerical_features if len(df_train[feature].unique())<25 and feature not in year_features+['Id']]
print("Discrete Variables in Numerical Variables Count: {} / {}".format(len(discrete_features),len(numerical_features)))

# discrete_features
```

```python
continuous_features=[feature for feature in numerical_features if feature not in discrete_features+year_features+['Id']]
print("Continuous features in Numerical Variables Count: {} / {}".format(len(continuous_features),len(numerical_features)))

# continuous_features
```

```python
# list of categorical variables
categorical_features=[feature for feature in df_train.columns if df_train[feature].dtypes=='O']
print("Categorical Variables Count: {}".format(len(categorical_features)))
```

### EDA Discrete Variables in Numerical Variables

```python
## Lets Find the realtionship between them and Sale Price

for feature in discrete_features:
    data=df_train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
## Also, lets look at the distribution of these variables, not just their relationship with the target
    
for feature in discrete_features:
    data=df_train.copy()
    data[feature].hist()
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

### EDA Continuous Variables in Numerical Variables

```python
## Lets analyse the continuous values by creating histograms to understand the distribution
## setting number of bins equal to 25, could add more if wanted more granular picture
for feature in continuous_features:
    data=df_train.copy()
    data[feature].hist(bins=100)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
# Outliers
for feature in continuous_features:
    
    data=df_train.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

### EDA Year Variables in Numerical Variables

```python
# let's explore the content of these year variables
for feature in year_features:
    print(feature, df_train[feature].unique())
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price

df_train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
## Here we will compare the difference between All years feature with SalePrice

for feature in year_features:
    if feature!='YrSold':
        df=df_train.copy()
        ## We will capture the difference between year variable and year the house was sold for
        df[feature]=df['YrSold']-df[feature]

        plt.scatter(df[feature],df['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

### Missing Values, errors, or outlier values

```python
# Show me the missing values and the propotion...
null_total = df_train.isnull().sum().sort_values(ascending=False)
null_percentage = (df_train.isnull().sum() / len(df_train)) * 100
missing_data = pd.concat([null_total, null_percentage], axis=1, keys=['Total', 'Percentage'])
only_missing_data = missing_data[missing_data['Percentage'] >= 0.01]
only_missing_data
```

```python
# How many feature has missing value more than 1%?
print('There are ' + str(only_missing_data.shape[0]) + ' variables out of 81 that have missing values.')
```

Let's find out the relationship between variables that has missing value with Sales Price

```python
for feature in list(only_missing_data.index):
     # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    df_train[feature] = np.where(df_train[feature].isnull(), 'missing', 'present')

    # let's calculate the mean SalePrice where the information is missing or present
    df_train.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```
